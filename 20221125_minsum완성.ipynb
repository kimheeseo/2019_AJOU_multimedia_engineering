# **SPA**

```python
def SPA(N, K, H, LLR, iteration) :
  message_matrix=np.zeros((K,N))
  message_chk_to_var=np.zeros((K,N)) #Message_CHK_to_VAR=zeros(K,N)
  LLR2=np.zeros((1,N)) #LLR 이전에 계산 된 값

  for i in range(0,N) :
    for j in range(0,K) :
        message_matrix[j,i]=LLR[i]
  A=np.zeros((N,3),dtype='int')
  B=np.zeros((K,6),dtype='int')

  for i in range(0,N):
    A[i,:]=A[i,:]+(np.where(H[:,i]==1)) # A{i}=find(H(:,i)==1);

  for j in range(0,K):
    B[j,:]=B[j,:]+(np.where(H[j,:]==1)) # B{j}=find(H(j,:)==1);
# A, B는 각각의 parity, information bit 의미

  BB=np.zeros((K, N))
  AA=np.zeros((K, B.shape[1]))
  AA2=np.zeros((N, A.shape[1]))
  AA3=np.zeros((1, N))
  AA4=np.zeros((1, N))
  MM=np.zeros((K,B.shape[1]))
  E=np.zeros((N, A.shape[1]))
  E1=np.zeros((N, A.shape[1]))

  for iiter in range(0,iteration):
    for j in range(0,K):
      for k in range(0, B.shape[1]) :
        AA[j, k]=message_matrix[j, B[j,k]] / 2  # B_j = B{j};
        AA[j, k]=np.tanh(AA[j, k]) # tanh(Message_matrix(j,B_j)/2)

      MM[j,:]=np.prod(AA[j, :],axis=0)/AA[j, :]
      aa=1+MM[j,:]
      bb=1-MM[j,:]
    
      for k in range(0, B.shape[1]) :
        bb[bb==0]=1e-12
        aa[aa==0]=1e-12
        message_chk_to_var[j,B[j,k]]=math.log(aa[k]/bb[k])
    #message_chk_to_var=np.clip(message_chk_to_var,-100,100)      

      mask1=message_chk_to_var>100
      message_chk_to_var[mask1]=100

      mask2=message_chk_to_var<-100
      message_chk_to_var[mask2]=-100
    AA3=np.zeros((1, N))
      
    for j in range(0, N) :
      for k in range(0, A.shape[1]) :
        AA3[0,j]+=message_chk_to_var[A[j,k],j]
      for k in range(0, A.shape[1]) :
        E[j, k]=AA3[0,j]-message_chk_to_var[A[j,k],j]
      for k in range(0, A.shape[1]) :
        message_matrix[A[j,k],j] = E[j,k] + LLR[j]
    AA4=np.zeros((1, N))
    for i in range(0,N) :
      for k in range(0, A.shape[1]) :
        AA4[0,i]+=message_chk_to_var[A[i,k],i]
      LLR2[0,i]=AA4[0,i]+LLR[i,0]
  return LLR2
```

# **MINSUM**

```python
def minsumdecoder(Li, H, K, N, iter):
  dc=6
  dv=3
  sys.float_info
  sys.float_info.max
  AA=np.zeros((K,N))
  BB=np.zeros((K, N))
  a=np.matlib.repmat(Li, K, 1)
  c=np.split(a,K,axis=0)
  c=np.array(c)
  c=c.reshape(K,N)

  CC=H*c
  zero=np.zeros((K,dc))

  for n in range(0,iter): #for n=1:iter
    #min sum
    #CC=H*c
    signvalue=np.sign(CC)
    absolutevalue=np.abs(CC)
    c1=np.argwhere(H==1)#c1=find(H(i,:));
    #c1은 H에서 1의 위치를 나타냄. matlab의 find와 동일       
    for j in range(0,1):
      for ii in range(0,dc):
        zero[j,ii]=c1[ii,1] 
      for ii in range(dc,dc*2):
        zero[j+1,ii-dc]=c1[ii,1] 

    for j in range(2,K):
      for ii in range(dc*j,dc*j+dc):
        zero[j,ii-dc*j]=c1[ii,1]             
    zero=np.asarray(zero, dtype = int)   #zero에 행은 x좌표를, 열은 y좌표를 나타낸다.  

    for i in range(0,K):#for i=1:K  
        #min value
      
      for k in range(0,dc): #for k=1:length(c1)
          minvalue=sys.float_info.max #minvalue=realmax;      
          for l in range(0, dc):
            if l !=k:
            #  print('absolutevalue.shape',absolutevalue)
              if absolutevalue[i,zero[i,l]]<minvalue:
                minvalue=absolutevalue[i,zero[i,l]]
              #  print('iteration값',n)
                #print(minvalue)
          CC=np.prod(signvalue[i,zero[i,:]])*signvalue[i,zero[i,k]];
          CC2=CC
          AA[i,zero[i,k]]=CC*minvalue;
  
      H2=H.T
      r1=np.argwhere(H2==1)
      zero2=np.zeros((N,dv))
      for j in range(0,1):
        for ii in range(0,dv):
          zero2[j,ii]=r1[ii,1] 

        for ii in range(dv,2*dv):
          zero2[j+1,ii-dv]=r1[ii,1]   

      for j in range(2,1008):
        for ii in range(dv*j,dv*j+dv):
          zero2[j,ii-dv*j]=r1[ii,1]             
      zero2=np.asarray(zero2, dtype = int)   #zero에 행은 x좌표를, 열은 y좌표를 나타낸다.    
      cHat=np.zeros((1,N)) 

      for j in range(0,N):  
        total=Li[j]+np.sum(AA[zero2[j,:],j])
          
        if total<0:
            cHat[0,j]=1
        else:
            cHat[0,j]=0
      CC=np.zeros((K,N))
      CC[0,0]=CC2
      for j in range(0,N):
        for k in range(0,dv):
          CC[zero2[j,k],j]=Li[j]+np.sum(AA[zero2[j,:],j])-AA[zero2[j,k],j]

      syndrome=np.mod(cHat.dot(H.T),2)
      np.sum(syndrome)

    if np.sum(syndrome)==0 :
        print('stop!!!')
        
        break;
  return cHat       
```
